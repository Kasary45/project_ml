```python
### In rent column data contain alphanumeric values like 1.5L,3.5L values we need to convert this into the numeric value
```

```python
## conversion of rentcolumn into numeri type:
def convert_rent(rent):
    if "L" in rent:
        return float(rent.replace('L','').replace(',','').strip())*100000                                           
    else:
        return float(rent.replace(',','').strip())                                          
                                                  

                                                  
                                                  
```

```python
city['Rent']=city['Rent'].apply(convert_rent)
```

```python
city.info()
```

```python
## conversion of Deposit column into the numeric type:
def convert_deposit(deposit):
    if deposit == "No Deposit":
        return 0
    elif isinstance(deposit, str) and 'L' in deposit:
        # Split on 'L', take the first part, convert to float, and multiply by 100000
        return float(deposit.split('L')[0]) * 100000
    else:
        # Handle other cases as needed
        return deposit

city['Deposit'] = city['Deposit'].apply(convert_deposit)



```

```python

## remove bathrooms from no_of bathroom:
def extract_first_number(x):
    if isinstance(x, str):
        # Split the string and convert the first part to float
        return float(x.split(' ')[0])
    elif isinstance(x, float):
        # If it's already a float, return it as is
        return x
    else:
        # Handle other cases (if any) or return NaN
        return float('nan')

city['Bathrooms'] = city['Bathrooms'].apply(extract_first_number)
```
