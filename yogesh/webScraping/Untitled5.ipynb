```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
import time
from bs4 import BeautifulSoup
import csv

# Function to save data to CSV file
def save_to_csv(data, filename):
    if data:
        keys = data[0].keys()
        with open(filename, 'w', newline='', encoding='utf-8') as csv_file:
            writer = csv.DictWriter(csv_file, fieldnames=keys)
            writer.writeheader()
            writer.writerows(data)
    else:
        print("No data to save.")

# Start the browser
service = Service(executable_path='/snap/bin/geckodriver')
browser = webdriver.Firefox(service=service)

# Open the webpage
browser.get("https://www.nobroker.in/property/sale/pune/Hinjawadi%20Village?searchParam=W3sibGF0IjoxOC41OTE3MzU3LCJsb24iOjczLjc0NTYyODA5OTk5OTk5LCJwbGFjZUlkIjoiQ2hJSklSbUFxdUc3d2pzUjJWTGEzcDhUaFcwIiwicGxhY2VOYW1lIjoiSGluamF3YWRpIFZpbGxhZ2UifV0=&radius=2.0&city=pune&locality=Hinjawadi%20Village")

# Wait for the page to load
time.sleep(5)  # Give the page some time to load

# Get the page source
page_source = browser.page_source

# Close the browser
browser.quit()

# Parse the HTML content
soup = BeautifulSoup(page_source, 'html.parser')

# Find all property containers
property_containers = soup.find_all('div', class_='nb__2JHKO')

# Initialize data list
data = []

# Loop through property containers
for container in property_containers:
    # Extract property description
    description = container.find('h2', class_='heading-6').text.strip()
    
    # Extract price per sqft
    price_per_sqft = container.find('div', class_='font-semi-bold').text.strip()
    
    # Extract property sqft
    property_sqft = container.find('div', class_='font-semi-bold').next_sibling.text.strip()
    
    # Extract property facing
    property_facing = container.find('div', class_='font-semi-bold', string='Facing').next_sibling.text.strip()
    
    # Append data to list
    data.append({
        'Description': description,
        'Price Per Sqft': price_per_sqft,
        'Property Sqft': property_sqft,
        'Facing': property_facing
    })

# Save the data to a CSV file
save_to_csv(data, 'property_data.csv')

print("Data has been saved to property_data.csv")
```
